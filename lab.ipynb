```python
import pandas as pd
import mysql.connector

# Koneksi ke database MySQL
db_connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='root',
    database='test',
    port=3307
)

# Membaca file CSV
data = pd.read_csv('data_covid.csv')

# Membuat kursor database
cursor = db_connection.cursor()


# Menyimpan data ke dalam database
for index, row in data.iterrows():
    date = row['Date']
    loc_iso = row['Location ISO Code']
    location = row['Location']
    cases = row['New Cases']
    deaths = row['New Deaths']
    
    # Query INSERT
    insert_query = f"INSERT INTO data_covid (`Date`, `Location ISO Code`, `Location`, `New Cases`, `New Deaths`) VALUES ('{date}', '{loc_iso}', '{location}', '{deaths}', '{cases}')"
    print(insert_query)
    cursor.execute(insert_query)

# Commit perubahan dan tutup koneksi
db_connection.commit()
cursor.close()
db_connection.close()

print("Data berhasil dimuat ke dalam database MySQL.")
```
